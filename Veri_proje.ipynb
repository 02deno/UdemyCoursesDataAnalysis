## Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
path_business = "3.1-data-sheet-udemy-courses-business-courses.csv"
path_all = "/content/Entry Level Project Sheet - 3.1-data-sheet-udemy-courses-web-development.csv"
```

```python
data_read = pd.read_csv(path_all)
```

```python
df=pd.DataFrame(data_read)
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

## Data Cleaning

```python
df.isnull().sum()  
```

```python
df.dropna(inplace=True)
```

```python
# Search for any duplicates 
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
# Convert published_timestamp datatype to datetime 
df['published_timestamp'] = df['published_timestamp'].str.replace("T", " ").str.replace("Z", "")
df.head()
```

```python
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
df.head(3)
```

```python
df['year_published'] = df['published_timestamp'].dt.year
df.head(3)
```

```python
del df['Date']
df
```

```python
df= df.sort_values('published_timestamp', ascending=False)
df
```

## Data Analysis

```python
level = df.groupby('level')['num_lectures'].count()
level
```

```python
subject = df.groupby('subject')['num_lectures'].count()
subject
```

```python
grouped = df.groupby('Free/Paid')['num_subscribers'].count()
grouped
```

```python
df.groupby('Free/Paid').mean()
```

```python
javascript_courses = df[df['course_title'].str.contains('JavaScript', na=False)]
len(javascript_courses)
```

```python
df.groupby(['level','subject']).size()
```

```python
# What are the five courses with the highest number of subscribers (num_subscribers column)?
course_sub = df.sort_values('num_subscribers', ascending=False).head(5)
course_sub
```

```python
meanPopRating = course_sub['rating'].mean()
print(f'The average rating among the most popular courses is {meanPopRating:.2f}')

maxPopRating = course_sub['rating'].max()
print(f'The maximum rating among the most popular courses is {maxPopRating}')

minPopRating = course_sub['rating'].min()
print(f'The minimum rating among the most popular courses is {minPopRating}')
```

```python
# best cources based on rating and number of subscribers
best_courses=df[(df['rating'] == df['rating'].max())].sort_values(by='num_subscribers' ,ascending = False).head()
best_courses
```

```python
# number of review in courses in all levels( that mean beginner level is most selling)
df.groupby('level')['num_reviews'].max().to_frame().sort_values('num_reviews',ascending = False)
```

```python
# number of subscribers in courses in all years
df.groupby('year_published')['num_subscribers'].max().to_frame().sort_values('num_subscribers',ascending = False)
```

## Data Visualization

```python
plt.figure(figsize=(16,8))
colors = ['#FFB9B2', '#A82D21', '#191919', '#DD4B3D']
ax = sns.barplot(x= level.index, y= level.values, palette = colors )
ax.set_title('Amount of Lectures per level', fontsize= 18)
for x, y in zip(range(0,len(level.index)), level.values):
    plt.annotate(str(y), xy = (x, y+10), fontsize = 16, backgroundcolor = 'grey', color = 'white', ha='center')
ax.set_yticks([])
ax.set_xlabel('Levels')
plt.xticks(fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(16,8))
ax = sns.barplot(x= subject.index, y= subject.values )
ax.set_title('Amount of Lectures per subject', fontsize= 18)
for x, y in zip(range(0,len(subject.index)), subject.values):
    plt.annotate(str(y), xy = (x, y+10), fontsize = 16, backgroundcolor = 'grey', color = 'white', ha='center')
ax.set_yticks([])
ax.set_xlabel('')
plt.xticks(fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
colors = ['#21618C', '#21618C', '#21618C', '#21618C']
df['subject'].value_counts().plot(kind='bar', color = colors)
plt.xticks(rotation=8)
plt.show()
```

### Which Course Subject Has The Highest Number Of Subscribers ?

```python
plt.figure(figsize=(8, 6))
colors = ['#FFB9B2', '#A82D21', '#191919', '#DD4B3D']
df.groupby(['subject'])['num_subscribers'].sum().plot(kind='pie', autopct='%1.0f%%', colors=colors)
plt.title('Distribution of subject by subscribers')
plt.show()
```

### Course With The Highest Number of subscribers

```python
top_ten_courses = df.sort_values(by='num_subscribers', ascending=False).head(10)
colors = ['#DD4B3D']
plt.figure(figsize=(8, 6))
sns.barplot(x='num_subscribers', y='course_title', data=top_ten_courses, palette = colors)
plt.title('Course With The Highest Number of subscribers', fontsize=(15))
plt.xlabel("Number Of Subscribers", fontsize=(13))
plt.ylabel("Course Title", fontsize=(13))
plt.show()
```
